```python
import pandas as pd
import numpy as np
from scipy.stats import linregress
import matplotlib.pyplot as plt

#Dados
preco_distribuicao = [54.86, 55.19, 54.58, 54.07, 53.72, 53.68, 54.29, 54.26, 53.92, 54.02, 55.78, 57.16]
preco_revenda = [71.40, 71.22, 71.56, 71.32, 71.27, 71.33, 71.46, 71.07, 71.68, 70.22, 73.68, 74.37]

#Criar DataFrame
df = pd.DataFrame({'PreçoDistribuicao': preco_distribuicao, 'PreçoRevenda': preco_revenda})

#Calculando a regressão linear
slope, intercept, r_value, p_value, std_err = linregress(preco_distribuicao, preco_revenda)

#Calculando a correlação
correlacao = df['PreçoDistribuicao'].corr(df['PreçoRevenda'])
print(f'Correlação entre Preço Distribuição e Preço Revenda: \n{correlacao}\n')

print(f'Correlação entre Preço Distribuição e Preço Revenda: \n{correlacao}\n')
print(f'Coeficiente angular (slope): {slope}')
print(f'Intercepto (intercept): {intercept}')
print(f'Coeficiente de determinação (R²): {r_value**2}')

x = np.array(preco_distribuicao)
y_tendencia = slope * x + intercept

plt.figure(figsize=(10, 6))
plt.scatter(preco_distribuicao, preco_revenda, color='blue', label='Dados Observados')
plt.plot(x, y_tendencia, color='red', label='Linha de Tendência')


plt.text(min(x) + 0.1, max(preco_revenda) - 0.1, f'y = {slope:.4f}x + {intercept:.4f}', fontsize=12, color='red')

plt.title('Regressão Linear entre Preço Distribuição e Preço Revenda')
plt.xlabel('Preço Distribuição (R$)')
plt.ylabel('Preço Revenda (R$)')
plt.legend()
plt.grid(True)
plt.show()

plt.plot(preco_distribuicao, preco_revenda)
plt.show
```
